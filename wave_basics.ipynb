```python
import librosa
#import pyAudio
import IPython.display as ipd
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
```

```python
file_path = \
    "/Users/nero/Splice/sounds/packs/Linden Jay So I Don't Forget Sample Pack/LINDEN_JAY_edited_labeled_processed/LINDEN_JAY_tonal/LINDEN_JAY_tonal_loops/"
file_name = \
    "LINDEN_JAY_keys_loop_flashlight_zelda_synth_92_Bbm.wav"
# synth
x, sample_rate = librosa.load(file_path + file_name)
print(x, sample_rate)

# drums
x2, sample_rate2 = librosa.load(
    "/Users/nero/Splice/sounds/packs/umru - sound cache vol 2/UMRU_labeled_processed/UMRU_drums/UMRU_drum_loops/UMRU_drum_loops_full/UMRU_drum_loop_restart_full_128.wav"
)
```

```python
# syth melody
ipd.Audio(x, rate=sample_rate)
```

```python
# display waveform in amplitude-time plot.
# This plot can show us the wave structure, so the energy (Amblitude) of a signal at certain timestep t.
# It cannot give us information about the frequency contained in the wave.
# It gives us information about how many notes are played in a certain timestep (bpm)
# maybe its measured by counting amplitude peaks
# gives us information whether the tone is hold or not, but not of how the tone would sound.
# Melodies for example have usually very short subsequent tones. A melody would encode
# in a sequence of amplitudes, where repetitive sample values (11111) mean the
# tone is hold for k timesteps.
# for example the shape of the decay (how the energy vanishes, or how the wave
# deconstructs itself in space or pressure calms dawn) indicates typical sounds.
# a synth-wave slowly decays whereas a snare-wave almost instantaneously disappears.

# Synth
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sample_rate)
plt.suptitle("Amplitude-Time Plot: Synth")

# Drums
plt.figure(figsize=(14,5))
ipd.Audio(x2, rate=sample_rate2)
librosa.display.waveplot(x2, sample_rate2)
plt.suptitle("Amplitude-Time Plot: Drums")
```
